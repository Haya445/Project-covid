```python
top_Recoverd=covid_19.groupby(by='Country/Region').max()[['Recovered']].sort_values(by=['Recovered'],ascending=False).reset_index()
fig=plt.figure(figsize=(18,5))
plt.title("Top  Country/Region with most Recoverd",size=25)
ax=sns.barplot(data=top_Recoverd.iloc[:12],y="Recovered",x="Country/Region",linewidth=2,edgecolor='red')
plt.xlabel("Country/Region")
plt.ylabel("Total Recoverd Cases")
plt.show()
```

```python
def visualizer_by_country (aspect, number_of_countries):
    grouped_country=covid_19.groupby('Country/Region')[str(aspect)]
    grouped_country.sum().sort_values(ascending=False)[:number_of_countries].plot(kind='bar')
    return grouped_country
```

```python
visualizer_by_country(aspect='Deaths', number_of_countries=5)
```

```python
visualizer_by_country(aspect='Confirmed', number_of_countries=5)
```

```python
clean_complete=pd.read_csv('covid_19_clean_complete.csv')
```

```python
clean_complete
```

```python
def visualize_month_by_month_country(country,aspect):
    df=clean_complete.drop(['Lat','Long','Active','Province/State'],axis=1)
    df2=df[df['Country/Region']==str(country)]
    df2=df2.reset_index()
    df2=df2.drop('index',axis=1)
    df2.Date=pd.to_datetime(df2.Date)
    df2['Month']=df2.Date.dt.month
    sns.barplot(x=df2.Month,y=df2[str(aspect)],color='gold')
    plt.xticks(ticks=[0,1,2,3,4,5,6,7],labels=['January','February','March','April','May','June','July','August'], rotation=90)
    plt.show()
```

```python
visualize_month_by_month_country(country='Saudi Arabia',aspect='Deaths')
```

```python
visualize_month_by_month_country(country='Saudi Arabia',aspect='Confirmed')
```
